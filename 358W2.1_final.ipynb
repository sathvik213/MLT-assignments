```python
import pandas as pd
import numpy as np
import statistics
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_excel('zomato.xlsx')
data=data.drop(columns=['City','Locality','Locality Verbose','Longitude','Latitude','Rating color','Cuisines','Address'],axis=1)
```

```python
data
```

```python
data.describe()
```

# Univariate Analysis 

## Box Plots 

```python

sns.boxplot(data=data)
```

```python
sns.boxplot(data=data['Aggregate rating'])
```

```python
sns.histplot(data=data['Price range'])
```

# Kernal Density Plots 

```python
sns.kdeplot(data['Aggregate rating'])
```

# Bivariate Analysis 

## Scatter Plot 

```python
sns.scatterplot(x=data['Restaurant ID'],y=data['Aggregate rating'])
warnings.filterwarnings('ignore')
```

```python
sns.barplot(data['Has Online delivery'],data['Price range'])
warnings.filterwarnings('ignore')
```

```python
sns.pairplot(data)
```

### Countplot 

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

sns.countplot('Has Table booking',data=data,ax=axes[0,0])
sns.countplot('Price range',data=data,ax=axes[0,1])
sns.countplot('Is delivering now',data=data,ax=axes[1,0])
sns.countplot('Switch to order menu',data=data,ax=axes[1,1])
warnings.filterwarnings('ignore')
```
