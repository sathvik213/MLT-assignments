```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('titanic.csv')
data
```

```python
data.shape
```

```python
male=data.loc[data.Sex=='male',:]
female=data.loc[data.Sex=='female',:]
print('Count of female : ',len(female))
print('Count of male : ',len(male))
male
```

```python
sur_male=data.iloc[list(data.Sex=='male') ,:]
sur_male_1=sur_male.iloc[list(sur_male.Survived==1),:]

sur_female=data.iloc[list(data.Sex=='female') ,:]
sur_female_1=sur_female.iloc[list(sur_female.Survived==1),:]


print('Count of female survived  : ',len(sur_female_1))
print('Count of male survived : ',len(sur_male_1))
print('Count of female non-survived : ',len(female)-len(sur_female_1))
print('Count of male non-survived : ',len(male)-len(sur_male_1))
```

```python
l=[1,2,3]
for i in l:
    pass_1=data.iloc[list(data.Pclass==i)]
    male_p1=pass_1.iloc[list(pass_1.Sex=='male')]
    male_p1_sur=male_p1.iloc[list(male_p1.Survived==1)]
    print('In class ',i,'male count: ',len(male_p1),'male survived in class',i ,': ',len(male_p1_sur))
    female_p1=pass_1.iloc[list(pass_1.Sex=='female')]
    female_p1_sur=female_p1.iloc[list(female_p1.Survived==1)]
    print('In class ',i,'female count: ',len(female_p1),'female survived in class',i ,': ',len(female_p1_sur))
    print()
```

```python
g_40=data.iloc[list(data.Age>40) ,:]
g_40_male=g_40.iloc[list(g_40.Sex=='male'),:]
g_40_female=g_40.iloc[list(g_40.Sex=='female'),:]

print('greater than 40 age people count: ',len(g_40))
print('greater than 40 age male count: ',len(g_40_male))
print('greater than 40 age female count: ',len(g_40_female))
print()
print('lesser than 40 age people count: ',len(data['Sex'])-len(g_40))
print('lesser than 40 age male count: ',len(data['Sex'])-len(g_40_male))
print('lesser than 40 age female count: ',len(data['Sex'])-len(g_40_female))
```
